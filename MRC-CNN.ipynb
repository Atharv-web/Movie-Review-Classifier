```python
import nltk
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
nltk.download('movie_reviews')
from nltk.corpus import movie_reviews
```

```python
# getting the docs and labels
# docs == after preprocessing will become the x dataset
# labels == after preprocessing will become the y dataset
docs,labels = [],[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        doc = ' '.join(movie_reviews.words(fileid))
        docs.append(doc)
        labels.append(1 if category == 'pos' else 0)
```

```python
# Preprocessing
max_words = 10000  # Set a maximum vocabulary size
tokenizer = Tokenizer(num_words=max_words,oov_token='<OOV>')
tokenizer.fit_on_texts(docs)
sequences = tokenizer.texts_to_sequences(docs)
padded_sequences = pad_sequences(sequences,maxlen=200,padding='post',truncating='post')
```

```python
# Splitting the dataset into train and test
x_train,x_test,y_train,y_test = train_test_split(padded_sequences,labels,test_size=0.2,random_state=42)
```

```python
# Converting the y_train, y_test lists to arrays
y_test = np.array(y_test)
y_train = np.array(y_train)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=max_words,output_dim=128,input_length=200),
    tf.keras.layers.Conv1D(128,5,activation='relu'),
    tf.keras.layers.GlobalMaxPooling1D(),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5,batch_size=64,validation_data=(x_test,y_test))
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
print(f"Test accuracy: {accuracy:.4f}")
```
