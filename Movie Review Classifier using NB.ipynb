```python
import nltk
from nltk import FreqDist, NaiveBayesClassifier
```

```python
nltk.download('movie_reviews')
```

```python
from nltk.corpus import movie_reviews
from nltk.classify import accuracy
import random
```

```python
docs = []
for categories in movie_reviews.categories():
  for fileids in movie_reviews.fileids(categories):
    doc = (list(movie_reviews.words(fileids)), categories)
    docs.append(doc)
```

```python
random.shuffle(docs)
```

```python
all_words = FreqDist(w.lower() for w in movie_reviews.words())
word_features = list(all_words)[:2000]
```

```python
# creating a function for

def document_feature(doc):
  doc_words = set(doc)
  features = {}
  for word in word_features:
    features['contains({})'.format(word)] = (word in doc_words)
    return features

feat_assets = [(document_feature(d),c) for (d,c) in docs]
```

```python
train_set, test_set = feat_assets[100:],feat_assets[100:]
```

```python
model = NaiveBayesClassifier.train(train_set)
```

```python
print(accuracy(model,test_set))
```
